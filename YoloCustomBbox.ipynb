```python
from ultralytics import YOLO
import supervision as sv
import cv2

model = YOLO("yolo26n.pt")
frame = cv2.imread("D:\Computer Vision\img1.jpg")
results = model.track(frame, persist=True)
```

```python

def display(
    frame,
    detections,
    labels,
    text_color=sv.Color.WHITE,
    text_position=sv.Position.TOP_LEFT,
    bbox_color=sv.Color.BLUE,
    thickness=1,
):
    bounding_box_annotator = sv.BoxAnnotator(color=bbox_color, thickness=thickness)
    label_annotator = sv.LabelAnnotator(
        color=bbox_color,
        text_color=text_color,
        text_position=text_position,
    )

    annotated_frame = bounding_box_annotator.annotate(
        scene=frame.copy(),
        detections=detections,
    )

    annotated_frame = label_annotator.annotate(
        scene=annotated_frame,
        detections=detections,
        labels=labels,
    )
    return annotated_frame

classId = results[0].boxes.id.int().cpu().numpy()
conf = results[0].boxes.conf.cpu().numpy()
xyxy = results[0].boxes.xyxy.cpu().numpy()

labels = [f"id:{int(_id)} p:{c:.2f}" for _id, c in zip(classId, conf)]

detections = sv.Detections(xyxy=xyxy, confidence=conf, class_id=classId)
x = display(frame, detections, labels)
```

```python
cv2.imshow("", x)
cv2.waitKey(0)   # chờ nhấn phím bất kỳ
cv2.destroyAllWindows()
```
